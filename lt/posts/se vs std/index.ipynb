## $\sigma$ vs SE 
*standartinis nuokrypis(standard deviation) vs. standartinė paklaida(standard error)*



Žymėjimai:

$X$ - atsititktinis dydis.
$$\mu = E[X] = \frac{1}{n}\sum_{i=0}^n(x_i)$$
$$Cov[X,Y] = E[(X-\mu_X)(Y-\mu_Y)]$$
$$Var[X]   = Cov[X,X] = E[(X - \mu)^2] = \frac{1}{n}  \sum_{i=0}^n(x_i - {\mu})^2$$


--------

Standartinis nuokrypis $\sigma = \sqrt{Var[X]}$, tai duomenų išsibarstymo matas(spread)skirstinio parametras.


Standartinė paklaida $SE = \frac{\sigma}{\sqrt{n}}$, tai išvestinio dydžio vidurkio standartinis nuokrypis.

----------
#### Pavyzdys:

Tarkim turim gazilijoną liniuočių.

Paimk atsitiktinai ir nepriklausomai 25 liniuotes,
išmatuojam, jog vidurkis lygus 20 cm, o standartinis nuokrypis lygus $\sigma$ = 1 cm.

---------
Esant gauso skirstiniui 68% duomenų patenka į $\pm 1$ cm intervalą aplink vidurkį. 

Arba yra 95% tikimybė, jog atsitiktinai pasirinkta liniuotė bus nutolusi nuo vidurkio ne daugiau nei 2 cm.

--------

Liniuočių vidurkio paklaida $\sqrt{25} = 5$, yra SE $= \frac{1}{5} = 0.2$ cm.


Išmatavę 100 liniuočių, tikėtina gausim labai artima vidurkį ~20cm,
 ir $\sigma$ beveik neturėtų pasikeisti toliau bus 1 cm.
 
Bet mūsų vidurkio paklaida $\sqrt{100} = 10$, bus mažesnė SE = 0.1cm.



$\sigma$ begaliniam duomenų limite priartėja konkrečią teigiamą vertę, bet SE tampa lygi 0.

----------

Kitas pavyzdys turime vieną liniuotę ir ją matuojame gazilijoną kartų.
Darydami prielaidą, jog neturime sisteminės paklaidos, liniuotės tikras ilgis bus vidurkis, o matavimai bus mūsų populiacija.

Ir nors matavimai skirsis, matuodami daug kartų eliminuosim atsitiktinę paklaidą.

Praktikoje, atsitiktinė paklaida matuojamo dydžio yra standartinė paklaida SE, dar padauginta stiudento koeficiento, todėl ją galima kiek norima sumažint eliminuoti išmatavus pakankamai daug kartų.



----
Nebūtinas normalusis skirstinys parametrams, pavyzdžiui, galima išmatuoti studentų amžių, 
cj standartinis nuokrypis ~5 metai. Bet vidurkis dėl studentų skaičiaus gali turėti labai mažą atsitiktinę paklaidą tarkim kelios dienos, dėl didelio studentų skaičiaus.


Beje jei po to matuosim visos žmonijos populiacijos amžių, tai gausim skirtingą ir standartinį nuokrypį ir vidurkį.
Nes studentų imtis nėra atsitiktinė(netolygiai reprezentuoja visą populiaciją.)
